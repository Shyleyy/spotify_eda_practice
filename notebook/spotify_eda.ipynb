```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv')

print(df.shape)

df.head(10)
```

```python
# drop the unnamed column
df.columns
df = df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
# checking data types
df.dtypes
```

```python
# Check for missing values
df.isna()

df.isna().sum()
```

```python
# check for missing values in the artists column
df.loc[df['artists'].isna()]
```

```python
# check for duplicates
df.duplicated(keep=False).sum()
```

```python
# check for duplicates in the track_id column
df.loc[df.duplicated(keep=False)].sort_values('track_id')
```

```python
# check for duplicates in the track_id column
df.loc[df['track_id'].duplicated(keep=False)].sort_values('track_id')
```

```python
# check the summary statistics of the dataset
df.describe()
```

```python
df['danceability'].hist(bins=30)
```

```python
# check for outliers in the danceability column
sns.boxplot(x=df['danceability'])
```

```python
df.columns
df_corr = df[['popularity',
       'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```

```python
# create a heatmap of the correlation matrix
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df_corr, annot=True, cmap='mako')
```

```python
sns.pairplot(df[['acousticness', 'energy']], height=5, aspect=1.5)
```

```python
# create a sample of the dataset to plot the pairplot
df.sample = df[['acousticness', 'energy']].sample(n=5000)
sns.pairplot(df.sample, height=5, aspect=1.5, diag_kind='kde')
```
